```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1, 11)
```

```python
y = 2 * x
```

```python
plt.title("2n Data")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
plt.plot(x,y, "ob")
plt.show()
```

### Quadratic Graph

**As the Definition, Quadratic Graph is ax^2 + bx + c**

and the graph will be curve

```python
x = np.arange(-99,100)
y = (x**2) + (6 * x) + 5
```

```python
plt.title("Quadratic Graph x^2 + 6x + 5")
plt.plot(x,y,"or")
plt.show()
```

```python
x = np.array([1, 2, 3])
y = np.array(x ** 2)

plt.bar(x, y, color = 'b', align = 'center')
plt.title("Bar Graph")
plt.show()
```
