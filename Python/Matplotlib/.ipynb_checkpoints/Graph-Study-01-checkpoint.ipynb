```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([x for x in range(100)])
y = np.array([y ** 2 for y in x])
```

```python
plt.title("Quadratic Graph")
plt.grid()
plt.plot(x, y, "r")
plt.show()
```

```python
x = np.array([a for a in range(20)])
y = np.array([(b ** 2 + 1) for b in x])
```

```python
plt.title("Quadratic Graph + 1")
plt.grid()
plt.plot(x, y, "or")
plt.show()
```
