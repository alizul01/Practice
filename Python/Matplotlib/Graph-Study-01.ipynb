```python
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

```python
x = np.array([x for x in range(100)])
y = np.array([y ** 2 for y in x])
```

```python
plt.title("Quadratic Graph")
plt.grid()
plt.plot(x, y, "r")
plt.show()
```

```python
x = np.array([a for a in range(20)])
y = np.array([(b ** 2 + 1) for b in x])
```

```python
plt.title("Quadratic Graph + 1")
plt.grid()
plt.plot(x, y, "or")
plt.show()
```

```python
# Graph of random numbers probabilty distribution
x = np.array([rd.randint(0, 10) for i in range(10)])
y = np.array([i for i in x])
```

```python
# Probability distribution of even numbers
plt.title("Even Numbers Probability Distribution")
plt.xlabel("Numbers")
plt.ylabel("Probability")
plt.grid()
plt.plot(x, y, "or")
plt.show()
```
