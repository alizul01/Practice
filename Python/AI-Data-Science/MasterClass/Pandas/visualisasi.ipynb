```python
import numpy as np
```

```python
np.random.seed(100)
```

```python
angkaRan = []
for i in range(0, 100):
    angkaRan.append(i)
```

```python
angkaBaru = pd.DataFrame(np.random.rand(100, 5), angkaRan, ['A', 'B', 'C', 'D', 'E'])
```

```python
angkaBaru.head()
```

```python
angkaBaru.plot.area(alpha = 0.6)
```
